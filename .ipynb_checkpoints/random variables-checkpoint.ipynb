### 1. Use the numerical variables from our dataset this week.

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
numerical=pd.read_csv('numerical.csv')
numerical
```

### 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
for col in numerical.columns:
    sns.displot(numerical[col],kde=True)
    plt.show()
```

From the above plots, we cant say for sure if the variables follow normal distribution. \
Also some variables need to be stretched to check if they follow normal distribution.

### 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

### Log Transformation

```python
num_transformed=np.where(np.isfinite(np.log10(numerical)),np.log10(numerical),0)
```

```python
#the result is an array so converting it to a datafranme for better readability and plotting

num_transformed=pd.DataFrame(num_transformed,columns=numerical.columns)
num_transformed
```

```python
for col in num_transformed.columns:
    sns.displot(num_transformed[col],kde=True)
    plt.show()
```

### Square Root Transformation
https://www.statology.org/transform-data-in-python/

```python
num_sqrt=np.sqrt(numerical)
num_sqrt
```

```python
for col in num_sqrt.columns:
    sns.displot(num_sqrt[col],kde=True)
    plt.show()
```

### Cubic Root Transformation

```python
num_cbrt=np.cbrt(numerical)
num_cbrt
```

```python
for col in num_cbrt.columns:
    sns.displot(num_cbrt[col],kde=True)
    plt.show()
```
